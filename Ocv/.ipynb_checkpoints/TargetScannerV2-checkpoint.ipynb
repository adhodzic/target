```python
import cv2
import numpy as np

image = cv2.imread('./images/targets/img6.jpg')
w = image.shape[1]
h = image.shape[0]

ratio = h/w
dim = (600, int(600*ratio))
image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)

imageOrig = image.copy()

bilateral = cv2.bilateralFilter(image,70,75,75)

bilateral = cv2.GaussianBlur(bilateral,(15,15),0)

# median = cv2.medianBlur(image,5)
# cv2.imshow('',median)


kernelSharpen = np.array([[-2,-2,-2],
                          [-2,17,-2],
                          [-2,-2,-2]])

bilateral = cv2.filter2D(bilateral,-1,kernelSharpen)

cv2.imshow('',bilateral)


# cv2.imshow('',bilateral)
cv2.waitKey()

lower = np.array([87, 0, 102])
upper = np.array([179, 255, 255])

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, lower, upper)

kernel1 = np.ones((5,5), np.uint8)

output = cv2.bitwise_and(image,image, mask= mask)


gray = cv2.cvtColor(output,cv2.COLOR_BGR2GRAY)
cv2.imshow('2',gray)
cv2.waitKey()
erosion = cv2.erode(gray,kernel1,iterations = 1)
dil = cv2.dilate(erosion,kernel1,iterations = 1)


edge = cv2.Canny(dil,50,255)
cv2.imshow('3',edge)
cv2.waitKey()

# img3 = cv2.cvtColor(output,cv2.COLOR_BGR2GRAY)

contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


def Pada(x1,y1,w,h,x,y,r):
    r = r/2
    if(x > x1+r and x < x1+w-r and
      y > y1+r and y < y1+h-r):
        return True
    else:
        return False

bg = np.zeros(gray.shape,dtype='uint8')
bg1 = np.zeros(gray.shape,dtype='uint8')
for c in contours:
    (x,y,w,h) = cv2.boundingRect(c)
    for i in range(len(c)):
        cont = c[i][0]
        bg1 = cv2.circle(bg1,(cont[0],cont[1]),16,30,2,8,0)
        bg = cv2.addWeighted(bg,1,bg1,1,0)
        bg1 = np.zeros(gray.shape,dtype='uint8')
    k, bg = cv2.threshold(bg, 220, 255, cv2.THRESH_BINARY)
    contoures, hierarchyy = cv2.findContours(bg, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    bg = cv2.medianBlur(bg,5)
    cv2.imshow('dete',bg)
    cv2.waitKey(0)
    for c in contoures:
        M = cv2.moments(c)
        if(M['m00'] != 0):
            cx = int(M['m10']/M['m00'])
            cy = int(M['m01']/M['m00'])
            if(Pada(x,y,w,h,cx,cy,16)):
                print('PADA')
                bilateral = cv2.circle(bilateral,(cx,cy),16,(255,0,0),1,8,0)
                bilateral = cv2.circle(bilateral,(cx,cy),1,(0,0,255),4,8,0)
    bg = np.zeros(gray.shape,dtype='uint8')
cv2.imshow('detected les',bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
